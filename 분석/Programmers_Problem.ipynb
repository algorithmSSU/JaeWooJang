## 등굣길

https://school.programmers.co.kr/learn/courses/30/lessons/42898

---

* 문제 요약 : 집에서 한 좌표까지의 경로의 경우의 수를 구하기  

* 문제 조건 : 물 웅덩이가 존재하며 물 웅덩이로는 지나갈 수 없다. 답은 1,000,000,007로 나눈 나머지를 반환한다.  

* 입력 조건 : 1 <= m, n <= 100, 0 <= Puddles Num <= 10  

---

* 사고의 흐름 :  
어떠한 위치에 도달할 수 있는 경우는 위에서 접근하는 경우와 좌측에서 접근하는 경우의 합이다. 물 웅덩이로 가는 경로는 존재하지 않고, 물 웅덩이에서 나아가는 경로 또한 존재하지 않는다. 이때 태두리를 0으로 가득 채워 범위에 대한 연산을 줄여주도록 하자.  
    
* 풀이 후 반성 :  
문제의 조건을 제대로 읽지 않아 1,000,000,007로 나누지 않아, 효율성 점검에서 계속 불합격 하여 필요 이상으로 시간을 낭비하였다.  


```python
int solution(int m, int n, vector<vector<int>> puddles) {
    int answer = 0;

    // 값을 기록할 matrix
    vector<vector<int>> mat(n + 2, vector<int>(m + 2, 0));

    // 물 웅덩이 저장
    for (int i = 0; i < puddles.size(); ++i)
        mat[puddles[i][1]][puddles[i][0]] = -1;

    mat[1][1] = 1;

    for (int i = 1; i <= n; ++i) {
        for (int j = 1; j <= m; ++j) {
            if (mat[i][j] == 0) {
                // 왼쪽에서 해당 칸에 접근 하는 경우의 수
                if (mat[i - 1][j] > 0)
                    mat[i][j] = (mat[i][j] + mat[i - 1][j]) % 1000000007;

                // 위쪽에서 해당칸에 접근 하는 경우의 수
                if (mat[i][j - 1] > 0)
                    mat[i][j] = (mat[i][j] + mat[i][j - 1]) % 1000000007;
            }
        }
    }

    answer = mat[n][m];

    return answer;
}
```

## 입국심사

https://school.programmers.co.kr/learn/courses/30/lessons/43238

---

* 문제 요약 : 처리 속도가 다른 t명의 입국심사원이 입국자 n명을 처리하는 최소 시간 반환  

* 문제 조건 : 마지막 사람의 입국 심사가 끝날때 까지의 시간  

* 입력 조건 : 1 <= peopleNum <= 1,000,000,000, 1 <= time <= 1,000,000,000, 1 <= officerNum <= 100,000  

---

* 사고의 흐름 :  
시간이 정해져 있다는 가정 하에 각 입국심사원들이 소화할 수 있는 사람의 수는 정해져 있다. 따라서 조건을 만족하는 최소한의 시간을 찾아가는 방식으로 이분 탐색적으로 접근한다. 이때 상한값은 가장 오래 걸리는 심사원이 n명 모두를 처리할때 걸리는 시간이다.   
    
* 풀이 후 반성 :  
형 변환을 신경쓰지 않은 탓에 오버플로우가 발생하여 불핗요한 시간을 많이 낭비하였다. 큰 값을 계산할땐 오버플로우 및 형 변환에 더 신경 써야 함을 다시 생각하게 되었다.


```python
long long solution(int n, vector<int> times) {
    long long answer = 0;

    int times_len = times.size();
    int max = 0;

    // 가장 오래 걸리는 시간 찾기
    for (int i = 0; i < times.size(); ++i)
        if (times[i] > max)
            max = times[i];

    long long left = 1;
    long long right = (long long)max * n;

    // 이분 탐색으로 최소 시간 구하기
    while (left <= right) {
        long long mid = (left + right) / 2;
        long long sum = 0;

        for (int i = 0; i < times_len; ++i)
            sum += mid / times[i];

        if (sum >= n)
            right = mid - 1;
        else
            left = mid + 1;
    }

    answer = left;

    return answer;
}
```

## 위장

https://school.programmers.co.kr/learn/courses/30/lessons/42578

---

* 문제 요약 : 옷을 입을 수 있는 경우의 수 구하기  

* 문제 조건 : 최소한 한 개 이상의 옷을 입어야 한다  

* 입력 조건 : 1 <= str_len <= 20  

---

* 사고의 흐름 :  
옷의 종류를 해시 함수로 숫자로 변환하여 같은지 여부를 빠르게 확인 할 수 있도록 한다. 이때 각 옷 종류의 개수를 카운트 하여 최종 결과를 계산 하도록 하자

```python

int solution(vector<vector<string>> clothes) {
    int answer = 1;

    hash<string> str_hash;
    vector<pair<int, int>> hash_clothes;

    for (const auto cloth : clothes) {
        int hash_val = str_hash(cloth[1]);

        int i;
        for (i = 0; i < hash_clothes.size(); ++i) {
            if (hash_clothes[i].first == hash_val) {
                // 옷의 종류가 같은 경우
                ++hash_clothes[i].second;
                break;
            }
        }

        if (i == hash_clothes.size())
            hash_clothes.emplace_back(hash_val, 1);
    }

    // 경우의 수 계산
    for (const auto hash_clothe : hash_clothes)
        answer *= (hash_clothe.second + 1);

    // 아무것도 입지 않는 경우 제외
    answer--;

    return answer;
}
```

## 베스트앨범(미해결)

https://school.programmers.co.kr/learn/courses/30/lessons/42579

---

* 문제 요약 : 장르 별로 많이 플레이 된 곡들로 이루어진 앨범 출력  

* 문제 조건 : 많이 재생된 장르 순서, 많이 재생된 노래 순서, 고유 번호가 낮은 노래 순서  

* 입력 조건 : 1 <= Play List Len <= 10,000, 1 <= Genre Num < 100  

---

* 사고의 흐름 :  
장르의 해쉬값과 새로운 클래스인 Album의 배열로 문제에 접근

```python
class Album {
public:
    int hash_val;
    int plays;
    vector<pair<int, int>> song;

    Album() {
        hash_val = 0;
        plays = 0;
    }

    Album(int hash_val, int plays, int i) {
        this->hash_val = hash_val;
        this->plays = plays;
        this->song.emplace_back(plays, i);
    }

    static bool cmpFunc(const Album& a, const Album& b) {
        return a.plays > b.plays;
    }
};

bool cmpFunc(const pair<int, int>& a, const pair<int, int>& b) {
    if (a.first > b.first)
        return true;

    else if (a.first < b.first)
        return false;

    return a.second < b.second;
}

vector<int> solution(vector<string> genres, vector<int> plays) {
    vector<int> answer;

    hash<string> str_hash;

    vector<Album> album_builder;
    for (int i = 0; i < genres.size(); ++i) {
        int hash_val = str_hash(genres[i]);

        int j;
        for (j = 0; j < album_builder.size(); ++j) {
            if (album_builder[j].hash_val == hash_val) {
                album_builder[j].plays += plays[i];
                album_builder[j].song.emplace_back(plays[i], i);
                break;
            }
        }

        if (j == album_builder.size())
            album_builder.emplace_back(hash_val, plays[i], i);
    }

    sort(album_builder.begin(), album_builder.end(), Album::cmpFunc);

    for (int i = 0; i < album_builder.size(); ++i) {
        sort(album_builder[i].song.begin(), album_builder[i].song.end(), cmpFunc);

        for (int j = 1; j < album_builder[i].song.size(); j += 2) {
            answer.emplace_back(album_builder[i].song[j - 1].second);
            answer.emplace_back(album_builder[i].song[j].second);
        }
    }


    return answer;
}
```
