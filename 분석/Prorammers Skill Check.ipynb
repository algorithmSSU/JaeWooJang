## 비밀지도

https://school.programmers.co.kr/learn/courses/30/lessons/17681

---

* 문제 요약 : 두 개의 숫자 배열을 복호화 하여 두 개의 지도를 만들고, 하나로 합쳐 출력  

* 문제 조건 : ' ' : 공간, '#' : 벽  

* 입력 조건 : 1 <= width <= 16, 0 <= x <=  2^n - 1  

---

* 사고의 흐름 :  
OR 연산을 이용해야겠다. 입력받은 숫자를 이진수로 바꾸고 바꾼 이진수를 or하도록 하자. 이때 숫자를 이진수로 바꾸는것이 생각보다 귀찮은 작업이라 생각 -> 숫자로 bit-연산자 or(|)를 먼저 하고 최종적인 숫자만 이진수로 바꾸어 1은 #, 0은 공백으로 출력  
    
* 풀이 후 반성 :  
숫자를 이진수로 변환하는 것을 머리로는 알고 있지만, 구현함에 있어서 깔끔하게 하지 못하고 어려움을 많이 겪었다. 아주 작은 기본기라도 갈고 닦아둬야함을 알게 되었다.    


```python
string binary(int n, int num) {
    string str(n, '#');

    int tmp = num;
    for (int i = n - 1; i >= 0; --i) {
        if (tmp % 2 == 0)
            str[i] = ' ';
        tmp *= 0.5f;
    }

    return str;
}

vector<string> solution(int n, vector<int> arr1, vector<int> arr2) {
    vector<string> answer;

    for (int i = 0; i < n; ++i) {
        int num = arr1[i] | arr2[i];
        answer.emplace_back(binary(n, num));       
    }

    return answer;
}
```

## K번째수

https://school.programmers.co.kr/learn/courses/30/lessons/42748

---

* 문제 요약 : 한 수열의 부분 수열을 정렬하였을 때 k번째 수 구하기  

* 문제 조건 : Command[] = { Begin, End, K'th }

* 입력 조건 : 1 <= Length <= 100, 1 <= Command Length <= 50  

---

* 사고의 흐름 :  
주어진 수열에서 [Begin:End]을 복사하고, 복사한 배열을 정렬한 이후 k번째를 출력하도록 하자  
    
* 풀이 후 반성 :  
부분 수열을 복사하는 것을 copy를 이용하였다면 더 좋았을 듯 하지만, 익숙하지 않은 함수였기에 당시에는 사용하지 못하고, for문을 이용해 배열을 새롭게 만들어 낸 점이 아쉬웠다.  

```python
vector<int> solution(vector<int> array, vector<vector<int>> commands) {
    vector<int> answer;
    for (int i = 0; i < commands.size(); ++i) {
        int left = commands[i][0] - 1;
        int right = commands[i][1] - 1;
        int k = commands[i][2] - 1;

        vector<int> tmp;
        for (int iter = left; iter <= right; ++iter)
            tmp.emplace_back(array[iter]);
        sort(tmp.begin(), tmp.end());

        answer.emplace_back(tmp[k]);
    }
    return answer;
}
```

## 124 나라의 숫자

https://school.programmers.co.kr/learn/courses/30/lessons/12899

---

* 문제 요약 : 3개의 숫자 1, 2, 4로 이루어진 숫자로 변환하기

* 문제 조건 :  1 -> 1, 3 -> 4, 4 -> 11, 5 -> 12, 9 -> 22 ...

* 입력 조건 : 1 <= n <= 500,000,000

---

* 사고의 흐름 :  
3진수와 상당히 유사하다. 모듈러 연산 및 나눗셈 연산으로 접근을 해보도록 하자. 하지만 단순한 3진수로는 접근할 수 없었다. 0이 존재하지 않았기 때문. 나머지 값을 바로 이용하기 보다는 { 4, 1, 2 } 배열을 만들고, 나머지 값의 인덱스로 결과를 출력하는 방식으로 접근한다. 이때 나머지가 0인 경우, 예외가 존재하여 그를 처리할 필요가 있었다.

* 풀이 후 반성 :  
예외가 존재한다는 것은 문제를 잘못 해석한 것이라 생각한다. 조금만 더 생각해보면 예외 케이스 없이 바로 결과를 만들어 내게끔도 가능할 것 같지만 깊게 생각해보고 싶지 않다.  

```python
string solution(int n) {
    string answer = "";

    char ch[3] = { '4', '1', '2'};

    stack<int> out;

    int rest = n;
    while (rest > 0) {
        int tmp = rest % 3;

        out.push(ch[tmp]);

        if (tmp == 0)
            rest -= 3;

        rest /= 3;
    }

    int i = 0;
    while (!out.empty()) {
        int tmp = out.top();
        out.pop();

        if (tmp != 3)
            answer.push_back(tmp);
        else
            answer.push_back(4);
    }

    return answer;
}
```

* 문제 요약 : 열쇠 배열로 자물쇠 배열을 겹치지 않고 가득 채울 수 있는지 여부 반환  

* 문제 조건 :   

* 입력 조건 : key는 M x M(3 <= M <= 20), lock은 N x N(3 <= N <= 20) 크기  

---

* 사고의 흐름 :  
Brute Force? 메트릭스를 돌리는 함수, 이동시키는 함수를 만들어서 변형을 하고, key와 lock을 xor 하여 해결 되는지 여부를 확인하도록 하자.  
잘못된 생각 : key와 lock의 사이즈가 같다 => 다르다
사이즈가 같다는 가정하의 xor는 이전에 이진수 -> 숫자변환 xor한 접근과 비슷하게 해보자는 생각을 하게됨. mat -> 행단위로 숫자로 변환, 변환된 숫자 xor 2^n -1 인지 확인

* 풀이 후 반성 :  
"제출 후 채점하기"이후 두 배열의 크기가 다를 수 있다는 사실을 인지하였으나, 알고리즘의 구현부를 상당수 수정해야 하였고, 시간이 부족한 관계로 포기하였다.

```python
bool rotation90(vector<vector<int>>& mat, int n) {
    vector<vector<int>> copyMat = copyMatrix(mat);
    for (int i = 0; i < n; ++i) {
        for (int j = 0; j < n; ++j) {
            mat[i][n - j - 1] = copyMat[j][i];
        }
    }

    return true;
}

bool move(vector<vector<int>>& mat, int x, int y) {
    vector<vector<int>> copyMat = copyMatrix(mat);

    int n = mat.size();
    for (int i = 0; i < n; ++i) {
        for (int j = 0; j < n; ++j) {
            if (((i + x) < n) && ((i + x) >= 0) &&
               ((j + y) < n) && ((j + y) >= 0))
                mat[i][j] = copyMat[i + x][j + y];
            else
                mat[i][j] = 0;
        }
    }

    return true;
}

bool check(vector<int>& key, vector<int>& lock) {
    int val1 = changeNum(key);
    int val2 = changeNum(lock);

    int tmp = val1 ^ val2;

    if (tmp != (pow(2, key.size()) - 1))
        return false;
    return true;

}

bool solution(vector<vector<int>> key, vector<vector<int>> lock) {
    int n = key.size();
    for (int rot = 0; rot < 4; ++rot) {
        rotation90(key, n);
        for (int i = -n + 1; i < n; ++i) {
            for (int j = -n + 1; j < n; ++j) {
                vector<vector<int>> copyMat = copyMatrix(key);
                move(copyMat, i, j);
                int k = 0;
                for (k = 0; k < n; ++k)
                    if (!check(copyMat[k], lock[k]))
                        break;

                if (k == n)
                    return true;
            }
        }
    }
    return false;
}
```
