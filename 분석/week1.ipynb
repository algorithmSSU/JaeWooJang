## 2156 | 포도주 시식

---

* 문제 요약 : 연속해서 최대 2개만 선택할 수 있을때 최대 합   

* 문제 조건 : 2초 | 128MB  

* 입력 조건 : 1 <= input len <= 10,000,  value >= 0    

---

* 사고의 흐름 :  
    이전의 결과가 앞으로에 영향을 줄 수 있을것으로 예상됨 -> DP로 접근  
    이전에 마신 경우 | 안마신 경우로 나누어 생각해 볼까?  
    => 10 10 1 1 10 10 의 경우를 커버 불가  
    
    DP적으로 생각해보았을 때 아무튼 이전 결과가 있고, 지금의 행동이 있을 것  
    지금 행동 { 마시지 않는다, 1개째 마신다, 2개째 마신다 }  
    => 해당 경우로 접근  
    
    "이전 결과" + "지금 행동"  
    #Case1 : max[n - 1] + "마시지 않는다"  
    #Case2 : max[n - 2] + "마시지 않았다(n - 1)" + "한잔 마신다"  
    #Case3 : max[n - 3] + "마시지 않았다(n - 2)" + "두잔 마신다"  
    
    => max(Case)  

* 풀이 후 반성 :  
    연속해서 3개까지 인줄 알았다. => 문제 잘 읽기  
    입력조건을 충분히 고려하지 않아 무한루프 발생  
    입력값 : 양수 -> 초기값 세팅 : 0 (x)  
    입력값 : 음수 -> 초기값 세팅 : -1(O)  

```python
...
int maxDrink(vector<int>& maxOut, vector<int>& input, int n) {
	if (n < 0)
		return 0;

	if (n < 2)
		return maxDrink(maxOut, input, n - 1) + input[n];

	if (maxOut[n] < 0) {
		vector<int> cases(3);
		int sum = 0;

		cases[0] = maxDrink(maxOut, input, n - 1) + 0;
		cases[1] = maxDrink(maxOut, input, n - 2) + 0 + input[n];
		cases[2] = maxDrink(maxOut, input, n - 3) + 0 + input[n] + input[n - 1];

		maxOut[n] = max(cases);
	}
			
	return maxOut[n];
}
...
```
