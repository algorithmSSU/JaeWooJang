## 11725 | 트리의 부모 찾기

---

* 문제 요약 : 루트 없는 트리가 주어지고 각 노드의 부모 구하기  

* 문제 조건 : 1초 | 256MB
  
* 입력 조건 : 2 <= N <= 100,000  

---

* 사고의 흐름 :   
트리가 이진 트리인가? 아니다 -> 1차원 배열 트리로 접근할 수 없다.  
=> set의 배열로 하여 연결된 노드들을 저장하는 방식으로, 루트부터 시작하며 자식들의 부모에 현재 index   

* 풀이 후 반성 : 
탐색을 위한 Queue에 자식들을 넣어야 했는데, 포문의 i를 넣어서 잘못된 탐색, 

```python
    ...

	vector<vector<int>> mat(n + 1, vector<int>());
	
	// 노드간 연결 입력 받기 -> set에 저장
	for (int i = 0; i < n - 1; ++i) {
		int vert1, vert2;

		cin >> vert1 >> vert2;

		mat[vert1].emplace_back(vert2);
		mat[vert2].emplace_back(vert1);
	}

	// answer : 해당 인덱스의 노드의 부모 값 저장
	vector<int> answer(n + 1, 0);
	queue<int> que;

	que.push(1);
	while (!que.empty()) {
		int current = que.front();
		que.pop();

		// 자식 idx에 현재 값(부모) 저장
		for (int i = 0; i < mat[current].size(); ++i) {
			// 이미 부모가 있는경우 <= 현재 값의 부모(자식X)
			if (!answer[mat[current][i]]) {
				que.push(mat[current][i]);
				answer[mat[current][i]] = current;
			}
		}
	}

    ...
```
