## 11047 | 동전 0

---

* 문제 요약 : 주어진 종류의 동전으로 일정 금액을 표현하기 위한 최소 갯수 구하기

* 문제 조건 : 1초 | 256MB | !!! 주어진 동전은 이전 동전의 배수 !!!

* 입력 조건 : 1 <= Kind Of Coin <= 10, 1 <= Money <= 100,000,000

---

#####  

* 사고의 흐름 :
    그리디 알고리즘을 사용한다.  
    가장 큰 단위부터 낮춰가며 최대 개수와 나머지를 구해 최소 개수를 구한다.  
    
#####  
    
* 풀이 후 반성 :
    문제 분석이 충분히 이루어 지지 않았다.
    1. 일반적인 동전 최소 개수 알고리즘은 그리디 알고리즘으로 해결 불가하다. (DP로 접근해야함)
    2. 하지만 문제 조건을 자세히 읽어보면 다음과 같은 조건이 있다.
    3. *A1 = 1, i >= 2인 경우에 Ai는 Ai-1의 배수*
    4. 해당 조건 덕분에 Greedy 알고리즘으로 접근이 가능한 것이다.
    
    이런 분석 없이 Greedy 챕터였기에 Greedy를 사용한 것은 큰 잘못
    
##### 
    
* Greedy 증명 :
    
    >Ai는 Ai-1의 배수 임으로 Ai = n * Ai-1로 표현할 수 있다.  
    money = ... + m * Ai-1 + ... 이라고 할때 만약 m이 n보다 클 경우  
    money = ... + (n + k) * Ai-1 ...  
    money = ... + n * Ai-1 + k * Ai-1 ...  
    money = ... + Ai + k * Ai-1 ... 로 표현 가능하다 (배수 였기에 가능한 것)  
    즉 개수를 m = n + k개에서 1 + k로 줄일 수 있다.  
        

```python
int problem11047() {
	int n, k;
	cin >> n >> k;

	vector<int> coinType(n);
	for (int i = 0; i < n; ++i)
		cin >> coinType[i];

	int count = 0;

	// 큰 단위의 돈을 우선적으로 지불한다.
	for (int i = n - 1; i > -1; --i) {
		count += k / coinType[i];
		k %= coinType[i];
	}

	cout << count;

	return 0;
}
```

## 1715 | 카드 정렬하기

---

* 문제 요약 : 여러 카드 집합 최소 횟수 합치기

* 문제 조건 : 2초 | 128MB

* 입력 조건 : 1 <= Card Set Num <= 100,000, 0 < Size of Card Set < 1,000

---

#####  

* 사고의 흐름 :

    이전의 Tim Sort에서의 Merge 최적화 부분에 해당하는 문제이다.  
    Point, 작은 것부터, 사이즈가 비슷한 그룹끼리 합치기 => 정렬 필요  
    이때 합쳐진 새로운 카드 집합 또한 다시 정렬 필요  
    => 우선순위 Queue를 이용하여 삽입, 삭제시마다 별도의 정렬 없이 자동 정렬  

```python
int problem1715() {
	int input;
	
	int n;
	cin >> n;

	// 오름차순 우선순위 큐
	priority_queue<int, vector<int>, greater<int>> arr;
	for (int i = 0; i < n; ++i) {
		cin >> input;
		arr.push(input);
	}

	int sum = 0;
	while (arr.size() > 1) {
		// 작은 두 수를 더하기
		int a = arr.top(); arr.pop();
		int b = arr.top(); arr.pop();

		// 더한 수를 다시 우선순위 큐에 추가
		arr.push(a + b);
		sum += a + b;
	}

	cout << sum;

	return 0;
}
```

## 1931 | 회의실 배정

---

* 문제 요약 : n개의 [시작:끝]시간이 주어질때 최대 사용 가능하능한 회의 개수구하기

* 문제 조건 : 2초 | 128MB

* 입력 조건 : 1 <= n <= 100,000, 0 <= Time <= 2^23 - 1

---

#####  

* 사고의 흐름 :
    사작시간과 끝나는 시간으로 pair객체 배열을 만들고, 정렬하여
    일찍 시작하는 순서로 뽑으면 되나?
    -> (1, 24), (2,3), (3,4), ... 만 생각해봐도 불가능하다.
    => 일찍 끝나는 순서로 뽑는다.
    
#####  
    
* 그리디 증명 :
    > [begin:end]의 시간의 범위가 주워졌다고 가정하자. 이때 [start:finish]를 회의의 
    시작과 끝이라고 하자 (start >= begin, finish <= end) 해당 회의를 진행 한 이후 시간의 
    범위는 [finish:end]이다. 즉 finish가 작을수록, [finish:end]의 폭이 넓어진다. 
    폭이 넓을수록 더 많은 회의를 선택할 기회가 생긴다. 따라서 조건을 만족하면서, 
    finish가 가장 작은 회를 선택하는 것이 최선이다. 


```python
int problem1931() {
	ios_base::sync_with_stdio(false);
	cin.tie(NULL);

	int count = 0;
	int bound = 0;

	int num;
	cin >> num;

	vector<pair<int, int>> arr;

	for (int i = 0; i < num; ++i) {
		int start, end;
		cin >> start >> end;
		// 끝시간과 시작시간을 배열에 삽입
		arr.emplace_back(end, start);
	}

	sort(arr.begin(), arr.end());

	for (int i = 0; i < num; ++i) {
		if (arr[i].second >= bound) {
			bound = arr[i].first;
			++count;
		}
	}

	cout << count;

	return 0;
}
```

## 1541 | 잃어버린 괄호

---

* 문제 요약 : 주어진 수식에 괄호를 추가하여 식의 결과가 최소가 되도록 선택

* 문제 조건 : 2초 | 128MB

* 입력 조건 : strlen <= 50, ch = { '0' ~ '9', '+', '-' }

---

#####  

* 사고의 흐름 :
    식이 최소가 되기 위해서 빼는 수가 많아야 한다. -부호가 나온 이후의 모든 수는 부호에 상관 없이 괄호를 통하여 빼기로 만들 수 있다. 따라서, -부호 이전 까지의 수를 더하고 -부호 이후의 수는 빼는 형식으로 답을 구한다.
#####  
    
* 풀이 후 반성 :
    코드의 가독성이 상당히 떨어진다. isdigit()함수를 이용하여 부호와 숫자 구별을 더 쉽게 표현 가능하였고, isBraket을 기준으로 나눈 부분에 주석이 더 추가 되면 좋을 듯 하다. 


```python
#include <iostream>
#include <cstring>
#include <vector>

using namespace std;

int problem1541() {
	char exper[51];
	int length;

	cin >> exper;
	length = strlen(exper);

	vector<int> numArr;

	int value = 0;
	int sum = 0;

	bool isBracket = false;
	for (int i = 0; i < length; ++i) {
		// Case : Operator
		if ((exper[i] == '+') || (exper[i] == '-')) {
			if (!isBracket) {
				sum += value;
				if (exper[i] == '-')
					isBracket = true;
			}
			else
				sum -= value;

			value = 0;
		}
		// Case : Value
		else
			value = value * 10 + exper[i] - '0';
	}
	if (!isBracket)
		sum += value;
	else
		sum -= value;

	cout << sum;

	return 0;
}
```
